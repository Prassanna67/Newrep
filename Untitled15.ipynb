```python
from google.colab import files

uploaded = files.upload()  # Upload multiple images manually
image_paths = list(uploaded.keys())  # Store filenames in a list
```

```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt

# Store results in a list
results = []

# Loop through each image and analyze gender & emotion
for image_path in image_paths:
    try:
        analysis = DeepFace.analyze(img_path=image_path, actions=['gender', 'emotion'])

        # Store the result in a dictionary
        for person in analysis:
            results.append({
                "image": image_path,
                "gender": person['dominant_gender'],
                "emotion": person['dominant_emotion']
            })
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
```

```python
def filter_results(gender=None, emotion=None):
    filtered = [res for res in results if
                (gender is None or res["gender"].lower() == gender.lower()) and
                (emotion is None or res["emotion"].lower() == emotion.lower())]

    if not filtered:
        print("No matching images found.")
        return

    for res in filtered:
        print(f"Image: {res['image']} | Gender: {res['gender']} | Emotion: {res['emotion']}")
        img = cv2.imread(res["image"])
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.show()

# Example Query: Find happy men
filter_results(gender="Man", emotion="angry")

# Example Query: Find angry women
filter_results(gender="Woman", emotion="happy")
```
